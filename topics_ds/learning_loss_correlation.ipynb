```python
correlation_history = pickle.load(open('ll4.0_n_hidden2_correlation_history.pkl', 'rb'))
```

```python
len(correlation_history)
```

```python
plt.plot(range(len(correlation_history)), [e[0] for e in correlation_history])
plt.xlabel('batch epoch')
plt.ylabel('spearman')
plt.show()
```

```python
plt.plot(range(len(correlation_history)), [e[1] for e in correlation_history])
plt.xlabel('batch epoch')
plt.ylabel('p value')
plt.show()
```

```python
print('Average spearman', sum([e[0] for e in correlation_history]) / len(correlation_history))
```

```python
print('Average p value', sum([e[1] for e in correlation_history]) / len(correlation_history))
```
